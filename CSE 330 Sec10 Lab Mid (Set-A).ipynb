Q1. Display a polynomial of degree 5 which is given below:

x^5 - 7x^4 + 12x^3 + x - 15

Also compute the approximate value of polynomial at x=2.


```python
#Q1
#Use built-in Polynomial class
coeff=[-15, 1, 0, 12, -7, 1]
p=Polynomial(coeff)
display(p)

x=2
value=0
for i in range(len(coeff)):
    value+=coeff[i]*(x**i)
print(value)
```

Q5. Implement this formula below:

```python
#Q5
#node_x means an array of nodes, x means a value, t means a value
def octane_basis(t, node_x, x):
  o_t=0 #initialization
  #Use a loop to find basis

  for j in range(len(node_x)):
    if (j) != (t):
        o_t +=  ((node_x[j] - node_x[t])/(node_x[j]-x))

  return o_t
```
